1. Contador de números pares e impares:


<br>
Escribe un programa que utilice un bucle for o while para contar y mostrar <br>
la cantidad de números pares e impares en un rango específico, <br>
por ejemplo, del 1 al 100.<br>


```python
cantidad_pares = 0
cantidad_impares = 0
lista_pares = []
lista_impares = []
limite = int(input("Dime el límite: "))
```

```python
for numero in range(1, limite + 1):
    if numero % 2 == 0:
        cantidad_pares += 1
        lista_pares.append(numero)
    else:
        cantidad_impares += 1
        lista_impares.append(numero)
```

```python
print("Cantidad de números pares:", cantidad_pares)
print("Cantidad de números impares:", cantidad_impares)
```

Lo mismo con while

```python
cantidad_pares = 0
cantidad_impares = 0
lista_pares = []
lista_impares = []
limite = int(input("Dime el límite :"))
```

```python
numero = 1
while numero <= limite:
    if numero % 2 == 0:
        cantidad_pares += 1
        lista_pares.append(numero)
    else:
        cantidad_impares += 1
        lista_impares.append(numero)
    numero += 1
```

```python
print("Cantidad de números pares:", cantidad_pares)
print("Cantidad de números impares:", cantidad_impares)
```

3. Tabla de multiplicar:


<br>
Pide al usuario que ingrese un número y luego muestra la tabla de <br>
multiplicar de ese número del 1 al 10 utilizando un bucle for.<br>


```python
numero = int(input("Ingrese un número: "))
```

```python
for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)
```
