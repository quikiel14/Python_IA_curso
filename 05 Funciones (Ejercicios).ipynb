2. Número Primo:


<br>
Escribe una función que determine si un número dado es primo o no. <br>
Pedirá al usuario que ingrese un número y muestra un mensaje <br>
indicando si es primo o no.<br>


```python
def numero_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

numero = int(input("Introduce un número para verificar si es primo: "))
if numero_primo(numero):
    print(f"{numero} es un número primo.")
else:    print(f"{numero} no es un número primo.")
```

```python
import math

def area_circulo(radio):
    return math.pi * radio ** 2

def area_cuadrado(lado):
    return lado ** 2

def area_triangulo(base, altura):
    return (base * altura) / 2

print("""
Elige la forma geométrica:
  1 = Círculo
  2 = Cuadrado
  3 = Triángulo
""")

try:
    opcion = int(input("Elige una opción (1-3): "))
    if opcion == 1:
        radio = float(input("Introduce el radio del círculo:"))
        print(f"El área del círculo es: {area_circulo(radio):.2f}")
    elif opcion == 2: 
        lado = float(input("Introduce el lado del cuardado"))
        print(f"el área del cuardado es: {area_cuadrado(lado):.2f}")
    elif opcion == 3:
        base = float(input("Introduce la base del triángulo: "))
        altura = float(input("Introduce la altura del triángulo: "))
        print(f"El área del triángulo es: {area_triangulo(base, altura):.2f}")
    else:
        print("Opción no válida. Por favor, introduce un número entre 1 y 3.")
except ValueError :
    print(f"Error: Debes introducir un número válido. No letras")

```

6. Fibonacci:


<br>
Implementa una función para generar los primeros n números de la <br>
secuencia de Fibonacci. Pide al usuario que ingrese el valor de n.<br>


```python
def fibonacci(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        secuencia = [0, 1]
        for i in range(2, n):
            secuencia.append(secuencia[-1] + secuencia[-2])
        return secuencia

n = int(input("¿Cuántos números de Fibonacci quieres generar? "))
print(f"\nLos primeros {n} números de Fibonacci son:")
print(fibonacci(n))
```

8. Factorial:


<br>
Crea una función para calcular el factorial de un número. <br>
Pide al usuario que ingrese un número y muestra el resultado.<br>


```python
def factorial(n):
    if n < 0:
        return "Error: no existe factorial de números negativos"
    if n == 0 or n == 1:
        return 1
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

n = int(input("Introduce un número: "))
print(f"El factorial de {n} es: {factorial(n)}")
```

9. Conversión de Temperatura:


<br>
Implementa funciones para convertir entre Celsius y Fahrenheit. <br>
Pide al usuario que ingrese la temperatura y la unidad, y luego <br>
realiza la conversión.<br>


```python
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_a_celsius(f):
    return (f - 32) * 5/9

print("""
¿Qué conversión quieres realizar?
  1 = Celsius → Fahrenheit
  2 = Fahrenheit → Celsius
""")

try:
    opcion = int(input("Elige una opción (1-2): "))

    if opcion == 1:
        temp = float(input("Introduce la temperatura en Celsius: "))
        print(f"{temp}°C = {celsius_a_fahrenheit(temp):.2f}°F")

    elif opcion == 2:
        temp = float(input("Introduce la temperatura en Fahrenheit: "))
        print(f"{temp}°F = {fahrenheit_a_celsius(temp):.2f}°C")

    else:
        print("Opción no válida.")

except ValueError:
    print("Error: debes introducir un número.")
```

10. Juego de Adivinanzas:


<br>
Desarrolla un juego simple en el que el programa elige un número aleatorio <br>
y el jugador tiene que adivinarlo. <br>
Proporciona pistas sobre si el número es mayor o menor. <br>
Utiliza funciones para organizar el código.<br>


```python
import random

def generar_numero():
    return random.randint(1, 100)

def pedir_intento():
    try:
        return int(input("Introduce tu intento: "))
    except ValueError:
        print("Error: debes introducir un número.")
        return pedir_intento()

def dar_pista(intento, numero_secreto):
    if intento < numero_secreto:
        print("Demasiado bajo, el número es mayor")
    elif intento > numero_secreto:
        print("Demasiado alto, el número es menor")
    else:
        return True
    return False

def jugar():
    numero_secreto = generar_numero()
    intentos = 0

    print("🎮 ¡Bienvenido al juego de adivinar el número!")
    print("Estoy pensando en un número entre 1 y 100...\n")

    while True:
        intento = pedir_intento()
        intentos += 1

        if dar_pista(intento, numero_secreto):
            print(f"\n ¡Correcto! El número era {numero_secreto}")
            print(f"Lo adivinaste en {intentos} intentos")
            break

    volver = input("\n¿Quieres jugar otra vez? (s/n): ")
    if volver.lower() == "s":
        jugar()

jugar()
```
