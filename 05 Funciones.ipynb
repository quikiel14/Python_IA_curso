**¿Qué es una función en Python?**<br>
Una función en Python es un bloque de código reutilizable que realiza una tarea <br>
específica. <br>
Puedes pensar en una función como una especie de máquina que toma ciertos <br>
valores de entrada, realiza un conjunto de acciones con esos valores y <br>
luego devuelve un resultado. <br>
Las funciones son una parte fundamental de la programación porque nos permiten <br>
dividir nuestro código en partes más pequeñas y manejables, lo que facilita la <br>
lectura, la depuración y la reutilización del código.<br>
**¿Por qué son importantes las funciones?**<br>
Las funciones son importantes por varias razones:<br>
1. **Reutilización de código:** Puedes escribir una función una vez y usarla <br>
muchas veces en tu programa. Esto ahorra tiempo y evita la repetición de código.<br>
2. **Facilitan la lectura:** Dividir tu programa en funciones más pequeñas hace <br>
que sea más fácil de entender y mantener. Cada función se enfoca en una tarea <br>
específica.<br>
3. **Modularidad:** Las funciones permiten dividir un programa en módulos más <br>
pequeños, lo que facilita el desarrollo colaborativo. Diferentes programadores <br>
pueden trabajar en funciones diferentes.<br>
4. **Depuración:** Si tienes un error en tu programa, las funciones te ayudan a <br>
aislar y solucionar problemas más fácilmente, ya que puedes probar cada función <br>
por separado.<br>

Función para comprobar si un número es par o impar

```python
def es_par(num: int) -> bool:
    """
    Entrada: num, número entero positivo. 
    Devuelve True si el número es par, 
    de otra manera devuelve False.
    """
    salida = (num % 2 == 0)
    return salida
```

Puedo ver el "return"

```python
es_par(124)
es_par(1987)
```

...o bien usarlo sin guardarlo.

```python
def imprime_si_es_par(num):
    if es_par(num):
        print("Es par")
    else:
        print("Es impar")
```

```python
imprime_si_es_par(129)
```

Función para multiplicar con dos argumentos<br>
En lugar de multiplicar, estamos sumando 'a', 'b' veces

```python
def multiplicacion(a, b):
    resultado = 0
    while b > 0:
        resultado += a
        #print(resultado, b)
        b -= 1
    return resultado
```

```python
multiplicacion(5, 80)
valor = multiplicacion(5, 20)
valor
```

Función que devuele el valor de una ecuación de 2ºgrado

```python
def evaluar_cuadratica(a, b, c, x):
    '''
    a, b, c: valores numéricos de los coeficientes de una ecuación
    de segundo grado
    x: valor de la variable x.
    '''
    solucion = a*x*x+b*x+c
    return solucion
```

```python
evaluar_cuadratica(2, 2, 3, 8)  # f(x=8) = 2x² + 2x + 3
```

Siempre los keyword arguments AL FINAL

Si quiero forzar que los argumentos sólo puedan introducirse por clave-valor<br>
puedo reescribir la función así (Sólo cambia el asterisco en los parámetros):

```python
def evaluar_cuadratica(*, a, b, c, x):
    '''
    a, b, c: valores numéricos de los coeficientes de una ecuación
    de segundo grado
    x: valor de la variable x.
    '''
    solucion = a*x*x+b*x+c
    return solucion
```

Si intentamos introducir ahora un argumento posicional obtendremos:


TypeError: evaluar_cuadratica() takes 0 positional arguments but 3 positional arguments (and 1 keyword-only argument) were given
<br>
# Error:<br>
evaluar_cuadratica(2, 2, 3, x=8)<br>
# Correcto:<br>
evaluar_cuadratica(x = 8, c = 3, a = 2, b = 2)<br>
# ¿Y si quiero calcular f(8) siendo f = x²+3?<br>
evaluar_cuadratica(a=1, c=3, x=8)<br>
# A los coeficientes puedo darles un valor "por defecto"<br>
# Pero los argumentos sin valor por defecto deben ir ANTES que los otros<br>
def evaluar_cuadratica_incompleta(x: float, a = 0, b = 0, c = 0) -> float:<br>
  


```python
    a, b, c: valores numéricos de los coeficientes de una ecuación
    de segundo grado
    x: valor de la variable x.
    '''
    solucion = a*x*x+b*x+c
    return solucion
```

```python
evaluar_cuadratica_incompleta()
evaluar_cuadratica_incompleta(3)
evaluar_cuadratica_incompleta(a=1, c=3, x=8)
evaluar_cuadratica_incompleta(a=2, b=2, c=3, x=8)
evaluar_cuadratica_incompleta(3, b=2)
```

¿Y si no sé el número de parámetros posicionales que voy a meter?

```python
def imprime_lineas(a, *lineas):
    print(a.upper())
    for arg in lineas:
        print(arg)
```

```python
num1 =3
imprime_lineas("Puedo imprimir", "las líneas", "que quiera", "fin", num1)
imprime_lineas("a")
```

Podría hacerlo con una lista

```python
def imprime_lineas_con_lista(lista):
    for linea in lista:
        print(linea)
```

```python
lista_a_imprimir = ("Puedo imprimir", "las líneas", "que quiera")
imprime_lineas_con_lista(lista_a_imprimir)
```

```python
imprime_lineas_con_lista(["Puedo imprimir", "las líneas", "que quiera"])
```

pero conviene saber lo que significa el asterisco porque aparece en la<br>
documentación de las funciones que importaremos

```python
def imprime_lineas2(**argumentos_por_clave):
    for clave, valor in argumentos_por_clave.items():
        print(f'{clave}= {valor}')
```

```python
imprime_lineas2(nombre='Juan', edad=30, ciudad='Madrid', calle = "calle")
```

Esto podríamos haberlo hecho con un diccionario

```python
def imprime_lineas_con_diccionario(diccionario):
    for clave, valor in diccionario.items():
        print(f'{clave}= {valor}')
        
diccionario = {"nombre":'Juan', "edad":30, "ciudad":'Madrid'}
imprime_lineas_con_diccionario(diccionario)
```

```python
imprime_lineas_con_diccionario(nombre='Juan', edad=30, ciudad='Madrid', calle = "calle")
```
