La sentencia `match` se utiliza para realizar diferentes acciones basadas en diferentes condiciones.

 La Sentencia Match en Python


<br>
En lugar de escribir muchas sentencias `if..else`, puedes usar la sentencia `match`.<br>
La sentencia `match` selecciona uno de muchos bloques de código para ser ejecutado.<br>
<br>
# Sintaxis


<br>
Así es como funciona:<br>
1. La expresión `match` se evalúa una vez.<br>
2. El valor de la expresión se compara con los valores de cada caso.<br>
3. Si hay coincidencia, se ejecuta el bloque de código asociado.<br>
El siguiente ejemplo usa el número del día de la semana para imprimir el nombre del día:<br>
<br>
# Ejemplo

```python
día = 4
match día:
    case 1:
        print("Lunes")
    case 2:
        print("Martes")
    case 3:
        print("Miércoles")
    case 4:
        print("Jueves")
    case 5:
        print("Viernes")
    case 6:
        print("Sábado")
    case 7:
        print("Domingo")
```

 Valor por Defecto


<br>
Usa el carácter guión bajo `_` como último valor de caso si quieres que un bloque de código se ejecute cuando no hay otras coincidencias:<br>
<br>
# Ejemplo

```python
día = 4
match día:
    case 6:
        print("Hoy es Sábado")
    case 7:
        print("Hoy es Domingo")
    case _:
        print("Esperando el fin de semana")
```


<br>
El valor `_` siempre coincidirá, por lo que es importante colocarlo como último caso para que funcione como caso por defecto.<br>
<br>
 Combinar Valores


<br>
Usa el carácter barra vertical `|` como operador "o" en la evaluación del caso para verificar más de un valor coincidente en un solo caso:<br>
<br>
# Ejemplo

```python
día = 4
match día:
    case 1 | 2 | 3 | 4 | 5:
        print("Hoy es día laboral")
    case 6 | 7:
        print("¡Me encantan los fines de semana!")
```

 Sentencias If como Guardias


<br>
Puedes agregar sentencias `if` en la evaluación del caso como una condición adicional:<br>
<br>
# Ejemplo

```python
mes = 5
día = 3
match día:
    case 1 | 2 | 3 | 4 | 5 if mes == 4:
        print("Un día laboral en Abril")
    case 1 | 2 | 3 | 4 | 5 if mes == 5:
        print("Un día laboral en Mayo")
    case _:
        print("Sin coincidencia")
```
