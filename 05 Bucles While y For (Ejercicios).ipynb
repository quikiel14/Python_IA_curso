1. Contador de números pares e impares:


<br>
Escribe un programa que utilice un bucle for o while para contar y mostrar <br>
la cantidad de números pares e impares en un rango específico, <br>
por ejemplo, del 1 al 100.<br>


## FOR 


```python
pares = 0
impares = 0
lista_pares = []
lista_impares = []
limite = int(input("Introduce el límite de números a evaluar: "))
for numero in range(1, limite + 1):
    if numero % 2 == 0:
        pares += 1
        lista_pares.append(numero)
    else:
        impares += 1
        lista_impares.append(numero)

print(f"Cantidad de números pares: {pares}")
print(f"Cantidad de números impares: {impares}")
print(f"Números pares: {lista_pares}")
print(f"Números impares: {lista_impares}")
```

2. Suma de números primos:


<br>
Crea un programa que solicite al usuario un número y utilice un bucle <br>
while para sumar todos los números primos menores o iguales al número <br>
ingresado.<br>


```python
del print
```

```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

numero = int(input("Introduce un número: "))

suma = 0
actual = 2
primos_encontrados = []

while actual <= numero:
    if es_primo(actual):
        suma += actual
        primos_encontrados.append(actual)
    actual += 1

print(f"Números primos encontrados:")
print(primos_encontrados)
print(f"Suma: {' + '.join(str(p) for p in primos_encontrados)} = {suma}")
```

3. Tabla de multiplicar:


<br>
Pide al usuario que ingrese un número y luego muestra la tabla de <br>
multiplicar de ese número del 1 al 10 utilizando un bucle for.<br>


```python
numero = int(input("Ingrese un número: "))
print(f"Tabla de multiplicar del {numero}:")

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

5. Adivina el número:


<br>
Crea un juego en el que el programa genera un número aleatorio y el <br>
usuario tiene que adivinarlo. Utiliza un bucle while para que el usuario <br>
pueda seguir intentando hasta que adivine el número. Proporciona pistas <br>
para indicar si el número a adivinar es mayor o menor que el intento del <br>
usuario.<br>


```python
from random import randint
```

```python
import random

numero_secreto = random.randint(1, 100)
intentos = 0

print("¡Bienvenido al juego de adivinar el número!")
print("Estoy pensando en un número entre 1 y 100...")

while True:
    intento = int(input("\nIntroduce tu intento: "))
    intentos += 1

    if intento < numero_secreto:
        print("Demasiado bajo, el número es mayor")
    elif intento > numero_secreto:
        print("Demasiado alto, el número es menor")
    else:
        print(f"¡Correcto! El número era {numero_secreto}")
        print(f"Lo adivinaste en {intentos} intentos")
        break
```
